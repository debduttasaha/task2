```python
#perform the clustering
data = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
```

```python
data.shape
```

```python
from sklearn.cluster import KMeans
dis=[]
k=range(1,10)
for i in k:
    algo=KMeans(n_clusters=i, n_init=100)
    algo.fit(data)
    dis.append(algo.inertia_)
```

```python
plt.plot(k,dis)
plt.title("Elbow method")
plt.show()
```

optimum value of k is 3

```python
dis
```

```python
algo1=KMeans(n_clusters=3)
algo1.fit(data)
```

```python
algo1.inertia_ #checking the sum of all the squares distance
```

```python
#centroid of all 3 clusters
algo1.cluster_centers_
```

```python
algo1.labels_
```

```python
sns.scatterplot(x=data["SepalLengthCm"], y=data["SepalWidthCm"], hue=algo1.labels_)

plt.scatter(algo1.cluster_centers_[:, 0], algo1.cluster_centers_[:,1], c = 'red', label = 'Centroids')
plt.show()
```
